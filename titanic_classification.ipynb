```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
```

```python
test_df
```

```python
# 先把答案取出來

y_train = train_df["Survived"]
```

```python
# 調整資料集：將訓練資料集的survive取出

x_train = train_df.drop(["Survived"], axis=1) # axis??
x_test = test_df
x_train
```

```python
# 將訓練資料集與測試資料集結合，為了方便填補或刪除缺失值（調整資料集）

total = pd.concat([x_train, x_test], ignore_index=True)
total
```

```python
# 確定有哪些欄位是有缺失的
total.isna().sum()
```

```python
# 補數字類型缺失值
# PClass理論上來說應該填最常出現，因為是類別

med = total.median()
med
```

```python
total = total.fillna(med)
total.isna().sum()
```

```python
# 補Embarked

most = total["Embarked"].value_counts().idxmax()
total["Embarked"] = total["Embarked"].fillna(most)
total.isna().sum()
```

```python
# 稍微整理，將遺失值太多的欄位或用不到的欄位刪除

total = total.drop(["Cabin", "Ticket", "PassengerId"], axis=1)
total
```

```python
# 將姓名分開

def namecut(n):
    return n.split(",")[-1].split(".")[0].replace(" ", "")
total["Name"].apply(namecut).value_counts()
```

```python
def namecut(n):
    mid = n.split(",")[-1].split(".")[0].replace(" ", "")
    if mid=="Mr" or mid=="Sir":
        return "Mr"
    elif mid=="Mrs" or mid=="Mme" or mid=="Lady":
        return "Mrs"
    elif mid=="Miss" or mid=="Mlle":
        return "Miss"
    elif mid=="Master":
        return "Master"
    else:
        return None
    
total["Name"] = total["Name"].apply(namecut)
total
```

```python
# 將欄位改成 One-Hot Encoding

total = pd.get_dummies(total)
total
```

```python
x_train_process = total.iloc[:train_df.shape[0]] # 只拿row：iloc
x_test_process = total.iloc[train_df.shape[0]:]
x_test_process
```
